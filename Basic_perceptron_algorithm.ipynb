```
w0 = np.random.normal()
```

```
def display(w, w0, X, Y):
  intercept= -w0 /w[1]
  slope = -w[0]/w[1]
  hplane_x = np.arange(0,10)
  hplane_y = slope * hplane_x + intercept
  plt.plot(X[:, 0][Y==-1], X[:, 1][Y==-1],'g^')
  plt.plot(X[:, 0][Y==1],X[:, 1][Y==1],'bs')
  plt.plot(hplane_x,hplane_y,'r-')
  plt.xlabel("X1")
  plt.ylabel("X2")
  plt.show()
```

```
display(w,w0,X,Y)
```

```
def learning_algo(w,w0,x,y):
  for iteration in range(1001):
    miss_classified = 0
    for i in range(len(x)):
      actual = np.sign(Y[i])
      half_space = np.sign(np.dot(w,x[i])+w0)

      if actual != half_space:
        miss_classified += 1
        w = w + actual * x[i]
        w0 = w0 + actual

    if iteration %500 == 0 :
      print("Iteration",iteration)
      display(w,w0,x,y)
```

```
learning_algo(w,w0,X,Y)
```
